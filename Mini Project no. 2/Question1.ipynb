# Activation functions

```python
def elu(x, alpha=1):
    return x if x > 0 else alpha * (np.exp(x) - 1)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

# Muculloch-pitts Neuron Function

```python
class McCulloch_Pitts_neuron():
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def model(self, x):
        if self.weights @ x >= self.threshold:
            return 1
        else:
            return 0
```

# Triangle Neurons

```python
def Area(x, y):
    neur1 = McCulloch_Pitts_neuron([2, -1], 2) # A to C
    neur2 = McCulloch_Pitts_neuron([-2, -1], -6) # A to B
    neur3 = McCulloch_Pitts_neuron([0, 1], 0) # B to C
    neur4 = McCulloch_Pitts_neuron([1, 1, 1], 3) # Second Layer

    z1 = neur1.model(np.array([x, y]))
    z2 = neur2.model(np.array([x, y]))
    z3 = neur3.model(np.array([x, y]))
    z4 = neur4.model(np.array([z1, z2, z3]))

    return list([z4]) # Output List
```

# Generate Random Values

```python
num_points = 2000
x_values = np.random.uniform(0, 4, num_points)
y_values = np.random.uniform(-1, 3, num_points)
```

# Classifying Values

```python
red_points = []
green_points = []

for i in range(num_points):
    z5_value = Area(x_values[i], y_values[i])
    if z5_value == [0]:
        red_points.append((x_values[i], y_values[i]))
    else:
        green_points.append((x_values[i], y_values[i]))

red_x, red_y = zip(*red_points)
green_x, green_y = zip(*green_points)
```

# Plots

```python
plt.figure(figsize=(8, 6))
plt.scatter(red_x, red_y, color='red', label='triangle_in = 0')
plt.scatter(green_x, green_y, color='green', label='tringle_out = 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('McCulloch-Pitts Neuron Outputs')

x_triangle = [1, 3, 2, 1]
y_triangle = [0, 0, 2, 0]
plt.plot(x_triangle, y_triangle, color='blue', label='Triangle')

plt.grid(True)
plt.legend(loc='upper right')
plt.xlim(0, 4)
plt.ylim(-1, 3)

plt.show()
```
